```python
def test(centerNm):
  URL=('http://apis.data.go.kr/B551172/Liver04/'
  'lvAlchbyType?'
  '&serviceKey=ODarp8sriuhcMZ%2FERCy11FLvPzQOYEZAa07KtRfmQK1%2FG0TUd1V0K9mUz01pzFjvYCGISFIVcjJdRYGRvhhv2w%3D%3D'
  '&numOfRows=100'
  '&pageNo=1'
  '&type=json'
  '&centerNm=국립암센터')

  response=requests.get(URL)
  r = response.json()
  items = r['response']['items']['item']

  ptAge=[item ['ptAge']for item in items ]
  critYr=[item ['critYr']for item in items ]
  ptSexCd=[item ['ptSexCd']for item in items ]
  statsTrgtNm=[item ['statsTrgtNm']for item in items ]

  centerNm=pd.Series(centerNm)
  critYr=pd.Series(critYr)
  ptSexCd=pd.Series(ptSexCd)
  statsTrgtNm=pd.Series(statsTrgtNm)

  df=pd.DataFrame({'나이':ptAge,'연도':critYr, '성별':ptSexCd, '음주여부':statsTrgtNm})
  return df
```

```python
iface = gr.Interface(test, gr.inputs.Dropdown(['국립암센터','삼성서울병원','연세암병원','아주대학교산학협력단','대구가톨릭대학교의료원','가천대길병원']), outputs=['dataframe'])
iface.launch(debug=True)
```
